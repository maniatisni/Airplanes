```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Airplane_Crashes_and_Fatalities_Since_1908.csv")
# Split Date into Month-Day-Year
data = data.join(data.Date.str.split(pat="/",expand=True)).rename(columns={0:"Month",1:"Day",2:"Year"})

# Split Time into Hours and Minutes
data = data.join(data.Time.str.split(pat=":",expand=True)).rename(columns={0:"Hour",1:"Minute"})

# Drop unecessary columns
data = data.drop(columns=['Flight #','Route','Registration','cn/In',2])

# Remove Hour rows not in the XX format.
data['Hour'] = data['Hour'].str.extract('(^\d{2}$)')
data.dropna(subset = ["Hour"], inplace=True)
```

```python
data['Hour'] = data['Hour'].astype(int)
data['Day'] = data['Day'].astype(int)
data['Month'] = data['Month'].astype(int)
```

```python
# I map Days/Hours/Month on a circle in order to represent the cyclical nature
# of the data and the fact that the 12th month has the same distance from
# month 1 and month 11. Same applies to hours and weekdays.
# Days of the week
data['Day_sin'] = np.sin((data.Day)*(2.*np.pi/7))
data['Day_cos'] = np.cos((data.Day)*(2.*np.pi/7))
# Hours
data['Hour_sin'] = np.sin(data.Hour*(2.*np.pi/24))
data['Hour_cos'] = np.cos(data.Hour*(2.*np.pi/24))
# Months - Since months in pandas begin from 1 and not 0
# I do "Month-1" in order to start counting from 0
data['Month_sin'] = np.sin((data.Month-1)*(2.*np.pi/12))
data['Month_cos'] = np.cos((data.Month-1)*(2.*np.pi/12))
```
